## **Brief Background**
### 1. What is the An/Ci curve and why do we care about it?  
During our lab practical # 1, you might already become acquainted with the use of a portable gas exchange system LiCOR-6800 to measure leaf gas exchange. One important type of gas exchange measurements is called the A<sub>n</sub>/C<sub>i</sub> curve, where A<sub>n</sub> is short for net assimilation rate of CO<sub>2</sub> and C<sub>i</sub> is short for intercellular CO<sub>2</sub> concentration. The example A<sub>n</sub>/C<sub>i</sub> curve is shown in **Figure 1** below. The A<sub>n</sub>/C<sub>i</sub> curve is so important because many important biotic related photosynthesis parameters (e.g. maximum Rubisco carboxylation rate, Vcmax, maximum rate of electron transport, Jmax, and day-time respiration rate, Rd) can be derived from the A<sub>n</sub>/C<sub>i</sub> curve. With these derived parameters, together with the environmental variables, we can use the Farquhar-von Caemmerer-Berry (FvCB) biochemical photosynthesis model (see our lecture # 2 for more details) to simulate plant ecophysiological response to current and future climate conditions.

### 2. FvCB photosynthesis model and associated two biochemical processes of leaf photosynthesis.  
As shown in our lecture #2, the FvCB photosynthesis model is the first well-accepted mechanistic model of leaf photosynthesis. In this model, the A<sub>n</sub> of a C3 plant leaf is hypothesized to be limited by the minimum rate of the two biochemical processes (see Eq. 1 below): (1) the net assimilation rate (A<sub>c</sub>) regulated by maximum Rubisco carboxylation rate, and (2) the net assimilation rate (A<sub>j</sub>) regulated by maximum rate of electron transport. Particularly, A<sub>c</sub> is also called the Rubisco-limited photosynthesis rate, and A<sub>j</sub> is also called RuBP regeneration-limited photosynthesis rate (see Fig. 1 and Slide #28 of lecture 2).  
$$ An = min \lbrace A_c, A_j\rbrace	\tag{Eq.1} $$  

## **Specific Aims**
We have two specific aims and two associated modeling scenarios for our lab practical #2, illustrated in more details as follows.  

**Aim 1:**  
We will show you how to use an [R model of leaf photosynthesis](https://remkoduursma.github.io/plantecophys/) (which is coded with the FvCB photosynthesis equations, and adapted from [Duursma, 2015](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0143346)) to extract important biotic related photosynthesis parameters from field-measured A<sub>n</sub>/C<sub>i</sub> curves. These important parameters include (1) Vcmax, Jmax, and Rd at the measurement temperature, and (2) Vcmax25, Jmax25, and Rd25 that re-scale the field-derived Vcmax, Jmax, and Rd to a reference temperature at 25 ‚ÑÉ.  

Two reasons associated with scaling these field-derived variables to 25 ‚ÑÉ include 1) all these variables are regulated by the underlying enzymes and thus show strong temperature-dependent response (see Slide #33 of lecture 3), and 2) by normalizing these variables to 25 ‚ÑÉ will make field-measurements that are undertaken at different measurement temperatures comparable, by which we can analyze the intraspecific and interspecific variations in these variables at the same reference temperature of 25 ‚ÑÉ.

To achieve our aim 1, the four A<sub>n</sub>/C<sub>i</sub> curves that were collected using a LiCOR-6400 portable gas exchange system from a tropical rainforest in Xishuangbanna, Yunnan, China by Dr. Wu's team in the summer of 2019 will be used in this lab. These four curves will include one used for lab demonstration, and the other three used for the lab assignment to examine whether you have learnt how to play with the model.

We will show you how to use the same model of leaf photosynthesis to simulate leaf photosynthesis response to current and future climate scenarios. As shown in our lecture 2, leaf photosynthesis is regulated jointly by biotic related parameters (e.g. Vcmax25, Jmax25, and Rd25) and environmental variables (e.g. leaf temperature, PAR, atmospheric CO<sub>2</sub> concentration, C<sub>a</sub>, and C<sub>i</sub>/C<sub>a</sub>).

```r
# set "Tcorrect = TRUE" "Patm = mean value of Atmospheric Pressure" for estimating Vcmax, Jmax at 25C
fit_at_25C <- fitaci(data, varnames = list(ALEAF = "An", Tleaf = "Tleaf", Ci = "Ci", PPFD = "PAR"), Tcorrect = TRUE, Patm = 101.3, citransition=647.5288)
print(fit_at_25C$citransition) # display Ci transition point between Ac and Aj
# Look at a summary of the fit
summary(fit_at_25C)
# Make a standard plot
plot(fit_at_25C)
```

```r
# set "Tcorrect = FALSE" "Patm = mean value of Atmospheric Pressure" for estimating Vcmax, Jmax at at measurement Temperature
fit_at_max <- fitaci(data, varnames = list(ALEAF = "An", Tleaf = "Tleaf", Ci = "Ci", PPFD = "PAR"), Tcorrect = FALSE, Patm = 101.3, citransition=647.5288)
print(fit_at_max$citransition) # display Ci transition point between Ac and Aj
# Look at a summary of the fit
summary(fit_at_max)
# Make a standard plot
plot(fit_at_max)
```

## Practice 2: Investigating Ecophysiological Responses to Environmental Drivers
Example: Analyzing how An, Vcmax, Jmax, and Rd vary with dynamic PAR, leaf temperature, and atmospheric CO2 concentration.

```r
# ********* Step 1: Set Model Parameters********************
# Setting physiological parameters and environmental drivers for simulating
Vcmax25 <- 95
Jmax25  <- 135
Rd25    <- -1.8
PAR     <- 1800
Patm    <- 101.3 # Atmospheric Pressure

# Set constant Ci or Dynamic Ci
# Ci  <- 345 # intercellular CO2 concentration
# Ca <- 420 # Atmospheric CO2 concentration
Ca <- seq(420, 420+80*3, by = 3*5)
Ci <- Ca*0.7
print(Ci)

# Set constant Tleaf or Dynamic Tleaf
# Tleaf  <- 28
Tleaf  <- seq(28, 28+0.05*80, by= 0.05*5)
print(Tleaf)

# ********* Step 2: Run the Model********************
# For Simulating An and Rd
res_photo <- Photosyn(Tleaf = Tleaf, PPFD = PAR, Ci = Ci, Patm = Patm, Vcmax = Vcmax25, Jmax = Jmax25, Rd0 = Rd25, TrefR = 25)
# For Simulating Vcmax and Jmax by Setting "returnParsOnly=TRUE"
res_physi <- Photosyn(Tleaf = Tleaf, PPFD = PAR, Ci = Ci, Patm = Patm, Vcmax = Vcmax25, Jmax = Jmax25, Rd0 = Rd25, TrefR = 25, returnParsOnly=TRUE)

# combine the results
res <- data.frame(
  An = res_photo$ALEAF,
  Rd = res_photo$Rd,
  Vcmax = res_physi$Vcmax,
  Jmax = res_physi$Jmax
)
print(res)
```

```r
# ********* Step 3: Before Plotting, Reminber to Run the Model ***************
options(repr.plot.width = 10, repr.plot.height = 5)

# Plotting Future An
years <- seq(2020,2100,by=5)
ggplot(res, aes(x = years)) +
  geom_line(aes(y = An, color = "An")) +
  labs(title = "An In the Future Climate Scenario 3",
       y = "Rate (Œºmol m-2 s-1)", color = "Parameter") +
  theme_minimal()
```

```r
# ********* Step 3: Before Plotting, Reminber to Run the Model ***************
options(repr.plot.width = 10, repr.plot.height = 5)

# Plotting An vs Tleaf
ggplot(res, aes(x = Tleaf)) +
  geom_line(aes(y = Vcmax, color = "An")) +
  labs(title = "Vcmax vs Temperature",
       y = "Rate (Œºmol m-2 s-1)", color = "Parameter") +
  theme_minimal()
```
